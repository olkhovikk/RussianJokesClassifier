<h3>Парсинг данных с сайта</h3>

Вытягиваю данные (текст анекдота, положительная оценка, отрицательная оценка, суммарная оценка) из сайта anekdot.ru. Посмотрела структуру сайта, вытягиваю с помощью поиска в div по классам **text** и **rates**.
Чтобы вытягивать по дням, которые в url сайта указаны '%Y-%m-%d', делаю цикл с изменением дат и указанным заранее годом.

```python
def generate_week_ranges(year):
    week_ranges = []
    for month in range(1, 13):
        start_date = datetime(year, month, 1)
        while start_date.month == month:
            if start_date.weekday() == 0:  # Если понедельник
                end_date = start_date + timedelta(days=6)  # Воскресенье
                week_ranges.append((start_date, min(end_date, datetime(year, month + 1, 1) - timedelta(days=1)) if month < 12 else end_date))
            start_date += timedelta(days=1)
    return week_ranges

def parse_aneks_by_week(week_ranges, base_url):
    jokes = []
    total_votes = []
    positive_votes = []
    negative_votes = []

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.5481.100 Safari/537.36"
    }

    for start_date, end_date in week_ranges:
        current_date = start_date
        while current_date <= end_date:
            url = f"{base_url}{current_date.strftime('%Y-%m-%d')}/"
            print(f"Парсинг URL: {url}")
            response = requests.get(url, headers=headers)

            if response.status_code != 200:
                print(f"Не удалось получить данные для {current_date.strftime('%Y-%m-%d')}")
                current_date += timedelta(days=1)
                continue

            soup = BeautifulSoup(response.content, 'html.parser')
            topic_blocks = soup.find_all("div", class_="topicbox")

            for topic in topic_blocks:
                text_block = topic.find("div", class_="text")
                if text_block:
                    jokes.append(text_block.get_text(strip=True))
                else:
                    jokes.append(None)

                rates_block = topic.find("div", class_="rates")
                if rates_block:
                    rates_data = rates_block['data-r'].split(';')
                    positive = int(rates_data[2])  # положительные голоса
                    negative = int(rates_data[3])  # отрицательные голоса
                    total = positive + negative   # сумма голосов
                    positive_votes.append(positive)
                    negative_votes.append(negative)
                    total_votes.append(total)
                else:
                    positive_votes.append(None)
                    negative_votes.append(None)
                    total_votes.append(None)

            current_date += timedelta(days=1)

    return pd.DataFrame({
        'joke': jokes,
        'total_votes': total_votes,
        'positive_votes': positive_votes,
        'negative_votes': negative_votes
    })

base_url = "https://www.anekdot.ru/release/anekdot/day/"
all_jokes = []

for year in [2022, 2023, 2024]: 
    print(f"\n Парсим анекдоты за {year} год")
    date_ranges = generate_week_ranges(year)
    data_year = parse_aneks_by_week(date_ranges, base_url)
    data_year["year"] = year
    all_jokes.append(data_year)

# склеиваем дф
joke_data = pd.concat(all_jokes, ignore_index=True)

joke_data.to_csv("joke_data_multi_year.csv", index=False, encoding="utf-8-sig")
```

<h3>Чистка перед эмбеддингом и кластеризацией</h3>

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

joke_data_clean = remove_outliers(joke_data, 'positive_votes')
joke_data_clean = joke_data_clean.reset_index(drop=True)
joke_data_clean = joke_data_clean[joke_data_clean['total_votes'] >= 10]
```

<h3>Авторазметка</h3>

```python
joke_data_label = joke_data_clean.copy().reset_index(drop=True)

def smart_label(row):
    pos = row["positive_votes"]
    neg = row["negative_votes"]
    total = row["total_votes"]

    if pd.isna(pos) or pd.isna(neg) or pd.isna(total) or total < 10:
        return 0  # мало голосов - плохой анек

    pos_ratio = pos / total
    pos_neg_ratio = pos / (neg + 1)

    if pos >= 70 and pos_ratio >= 0.65 and pos_neg_ratio > 3:
        return 2  # гениальный
    elif pos >= 30 and pos_ratio >= 0.4 and pos_neg_ratio > 1.2:
        return 1  # норм
    else:
        return 0  # слабый

# размечаем
joke_data_label["label"] = joke_data_label.apply(smart_label, axis=1)
joke_data_label.to_csv("joke_data_autolabeled.csv", index_label="id", encoding="utf-8-sig")
```

<h3>Генерирую эмбеддинги</h3>

Загружаю предобученные токенизатор и модель Tiny-BERT на русском языке. 

Токенизирую текст, чтобы подготовить его для модели. Для модели нужны тексты одинаковой длины символов. Он обрезает текст длиннее 48 символов, и добавляет спец символ [PAD] в текст, короче 48 символов

Пропускаю токенизированный текст через модель и получаю эмбеддинги

```python
joke_data = joke_data_label

# TinyBERT - оптимизированный
tokenizer = BertTokenizer.from_pretrained("cointegrated/rubert-tiny") # предобученный токенизатор (русский)
model = BertModel.from_pretrained("cointegrated/rubert-tiny") # предобученная модель русский tinybert

# для создания эмбеддингов
def create_embeddings(texts, batch_size=32, max_len=48):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        inputs = tokenizer(batch_texts.tolist(), return_tensors="pt", truncation=True,
                           padding=True, max_length=48)
        with torch.no_grad():
            outputs = model(**inputs)
        batch_embeddings = outputs.last_hidden_state.mean(dim=1).detach().numpy()
        embeddings.append(batch_embeddings)
    return np.vstack(embeddings)

# тренировка + тест на основе авторазметки
jokes_train, jokes_test, ratings_train, ratings_test = train_test_split(
    joke_data['joke'],
    joke_data['label'],
    test_size=0.2,
    random_state=42,
    stratify=joke_data['label']  # баланс классов
)

# генерация эмбеддингов
train_embeddings = create_embeddings(jokes_train)
test_embeddings = create_embeddings(jokes_test)
```

<h3>Балансировка и кластеризация</h3>

Чтобы получить сбалансированные данные в каждом кластере использую RandomUnderSampler. Это метод балансировки классов в наборе данных, который уменьшает количество примеров в доминирующих классах. Делает он это случайным выбором.

Данные были масштабированы перед кластеризацией с использованием **StandardScaler**. Масштабирование данных привело их к стандартному виду - с нулевым средним значением и стандартным отклонением, равным единице. Это помогает уравновесить влияние всех признаков и улучшает производительность алгоритмов кластеризации.

Для кластеризации был использован алгоритм **MiniBatchKMeans** - оптимизированная версия KMeans, которая использует мини-батчи данных для обновления центроидов кластеров, что делает его более быстрым и эффективным при работе с большими наборами данных. *n_clusters=3* указывает на количество кластеров, которые должны быть выделены (плохой анекдот, хороший и отличный). *init='k-means++'* используется для улучшенной инициализации центроидов, что позволяет ускорить процесс и обеспечить лучшую сходимость.

<h5>Вариант 1. Используя UnderSampler</h5>

RandomUnderSampler хоть и выравниет классы, но смазывает картину их глобального распределения в выборке, из-за чего модель может переоценивать наименее выраженный класс. К тому же, в данном примере более 30 тысяч анекдотов потеряются, что сильно сократит выборку.

```python
# undersampling
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(train_embeddings, ratings_train)

# масштабирование перед кластеризацией
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# MiniBatchKMeans
kmeans = MiniBatchKMeans(n_clusters=3, init='k-means++', random_state=42)
labels = kmeans.fit_predict(X_scaled)

cluster_names = {
    2: "гениальный",
    1: "неплохо, можно лучше",
    0: "плохо, не пытайся в юмор",
}

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

df_resampled = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])
df_resampled['cluster'] = labels
df_resampled['cluster_name'] = df_resampled['cluster'].map(cluster_names)
df_resampled['positive_votes'] = y_resampled

# визуализация кластеров 
plt.figure(figsize=(12, 8))
for cluster_id, cluster_name in cluster_names.items():
    subset = df_resampled[df_resampled['cluster'] == cluster_id]
    plt.scatter(subset['PC1'], subset['PC2'], label=cluster_name, alpha=0.5)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.title('PCA кластеры')
plt.savefig("figures/PCA_balanced.png") 
plt.show()
```

<h5>Вариант 2. Не используя UnderSampling (делаю этот вар)</h5>

В этом варианте идет нормальзация эмбеддингов, а этап балансировки классов переносится на момет обучения моделей. 

```python
# масштабирование
scaler = StandardScaler()
X_scaled = scaler.fit_transform(train_embeddings)

# кластеризация
kmeans = MiniBatchKMeans(n_clusters=3, init='k-means++', random_state=42)
labels = kmeans.fit_predict(X_scaled)

# кластеры
cluster_names = {
    2: "гениальный",
    1: "неплохо, можно лучше",
    0: "плохо, не пытайся в юмор"
}

# PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

df_viz = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])
df_viz['cluster'] = labels
df_viz['cluster_name'] = df_viz['cluster'].map(cluster_names)
df_viz['true_label'] = ratings_train.reset_index(drop=True)

plt.figure(figsize=(12, 8))
for cluster_id, cluster_name in cluster_names.items():
    subset = df_viz[df_viz['cluster'] == cluster_id]
    plt.scatter(subset['PC1'], subset['PC2'], label=cluster_name, alpha=0.5)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.title('PCA кластеры')
plt.savefig("figures/PCA_wo_balanced.png") 
plt.show()
```

Заметно, что обнаружение классов без балансировки более четкое. Вариант норм работает

<h3>Обучаем наши модели</h3>

Были построены три модели. Логистическая регрессия - линейная модель для бинарной и многоклассовой классификации. Основана на логистической функции для предсказания вероятностей. SVM (Support Vector Machine) - метод, который находит гиперплоскость, максимально разделяющую классы. MLP (Multilayer Perceptron) - вид нейронной сети, способный моделировать нелинейные зависимости. Состоит из нескольких слоев нейронов (входной, скрытые, выходной)

```python
from sklearn.utils.class_weight import compute_class_weight


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, labels, test_size=0.2, random_state=42, stratify=labels
)

models = {
    'Logistic Regression': LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000),
    'SVM': SVC(random_state=42, class_weight='balanced', probability=True),
    'MLP': MLPClassifier(random_state=42, max_iter=500) 
}


# обучение
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1
    }

print("Сравнение моделей:")
for name, metrics in results.items():
    print(f"\n{name}:")
    for metric_name, metric_value in metrics.items():
        print(f"{metric_name} = {metric_value:.3f}")

# лучшая
best_model_name = max(results, key=lambda x: results[x]['F1-Score'])
best_model = models[best_model_name]
y_pred_best = best_model.predict(X_test)

print(f"\nОтчет по {best_model_name}:\n")
print(classification_report(y_test, y_pred_best, zero_division=0))
```

Для уверенности в правильности была проведена кросс-валидация.

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer

models = {
    'Logistic Regression': LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000),
    'SVM': SVC(random_state=42, class_weight='balanced'),
    'MLP': MLPClassifier(random_state=42, max_iter=500)
}

scoring = {
    'accuracy': make_scorer(accuracy_score),
    'f1_weighted': make_scorer(f1_score, average='weighted'),
    'precision_weighted': make_scorer(precision_score, average='weighted', zero_division=0),
    'recall_weighted': make_scorer(recall_score, average='weighted', zero_division=0),
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {}

for name, model in models.items():
    print(f"Кросс-валидация - {name}")
    scores = {}
    for score_name, scorer in scoring.items():
        cv_scores = cross_val_score(model, X_scaled, labels, cv=cv, scoring=scorer)
        scores[score_name] = {
            'mean': cv_scores.mean(),
            'std': cv_scores.std()
        }
        print(f"{score_name}: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
    results[name] = scores
    print("-" * 40)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def detect_topic(text):
    text = text.lower()
    for topic, keywords in topics.items():
        if any(word in text for word in keywords):
            return topic
    return 'Другое'

joke_data['topic'] = joke_data['joke'].apply(detect_topic)

topic_counts = joke_data['topic'].value_counts().to_dict()

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='magma'
).generate_from_frequencies(topic_counts)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Облако слов топиков", fontsize=16)
plt.tight_layout()
plt.savefig("figures/Topic_word_cloud.png") 
plt.show()
```

<h3>Проверка качества твоей шутки</h3>

```python
tokenizer = BertTokenizer.from_pretrained("cointegrated/rubert-tiny")
bert_model = BertModel.from_pretrained("cointegrated/rubert-tiny")

def predict_cluster_for_joke(joke_text, tokenizer, bert_model, scaler, logistic_model):
    inputs = tokenizer([joke_text], return_tensors="pt", truncation=True, padding='max_length', max_length=48)
    with torch.no_grad():
        outputs = bert_model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).detach().numpy()

    scaled = scaler.transform(embedding)

    cluster = logistic_model.predict(scaled)[0]

    labels = {
        0: "Плохо, не пытайся в юмор",
        1: "Неплохо, можно лучше",
        2: "Гениальный анекдот! Браво"
    }

    return labels.get(cluster, f"кластер {cluster}")
```

```python
anekdots = [
    "В семье Райана Гослинга дети с нетерпением ждали, когда отец скажет свое первое слово",
    "Почему курица перешла дорогу? Потому что она неудачник.",
    """
Стоит феминистка и громко ржёт. Подходит вторая:
— Что случилось?
— Да анекдот рассказали очень смешной. Только он про феминисток и очень обидный.
— Ты расскажи, только замени феминисток на мужиков.
Первая рассказала. Вторая долго ржет и говорит:
— Эти мужики пиздец тупые"
    """,
    "Сел медведь в машину и сгорел"
]

for joke in anekdots:
    print(joke)
    print("→", predict_cluster_for_joke(joke, tokenizer, bert_model, scaler, logistic_model))
    print()
```
